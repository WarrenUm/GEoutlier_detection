```python
import pandas as pd
from osrs_fcns import *
import plotly.express as px
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output, State
```

```python
app = dash.Dash()
```

```python
app.layout = html.Div([
    html.H1("Item Plotter",style={'text-align':'center'}),
    
#     dcc.Input(id='usrInput',placeholder='Enter a value...',
#         type='text',
#         value=''),
    
    html.Div(dcc.Input(id='input-box', type='text')),
    html.Button('Submit', id='button-example-1'),
#     html.Div(id='output-container-button',
#              children='Enter a value and press submit'),
    
    
    html.Br(),
    dcc.Graph(id='timeSeriesGraph',figure={})
    
    
])

@app.callback(
[Output(component_id='timeSeriesGraph',component_property='figure'),
 Input('button-example-1', 'n_clicks'),
 State('input-box', 'value')])
# Output('output-container-button', 'children'),
 #Input(component_id='usrInput',component_property='value')

def updateGraph(n_clicks,value):
    print(value)
    
    #get the data limit right now
    itemDF = resampleDF(getItemPrices([value],'03/08/21 00:00:00', '03/09/21 00:00:00'),'6h')
    
    fig = px.scatter(x=itemDF.index,y=itemDF.avgHighPrice,template='plotly_dark')
    
    return fig
```
