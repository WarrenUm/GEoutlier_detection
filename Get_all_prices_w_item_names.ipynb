```python
#fetching data from the db and formatting it into a dataframe. Also merging on the item id to get the name.
```

```python
import sqlite3
from sqlite3 import Error
import pandas as pd
```

```python
def get_items():
    conn = sqlite3.connect('geitems.db')
    cur = conn.cursor()
    cur.execute("""SELECT item_id, name FROM items""")
    response = cur.fetchall()
    df = pd.DataFrame(response,columns=['item_id','name'])
    conn.close()
    return df
```

```python
df = get_items()
```

```python
def get_prices():
    conn = sqlite3.connect('geitems.db')
    cur = conn.cursor()
    cur.execute("""SELECT * FROM prices""")
    response = cur.fetchall()
    df = pd.DataFrame(response,columns=['id','item_id','timestamp','avgHighPrice','highPriceVolume','avgLowPrice','lowPriceVolume'])
    conn.close()
    return df
```

```python
price_df = get_prices()
```

```python
def join_item_names(item_df,price_df):
    joined_df = item_df.join(price_df,on='item_id')
    return joined_df
```
